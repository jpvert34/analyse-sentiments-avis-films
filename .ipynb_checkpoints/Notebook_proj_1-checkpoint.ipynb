# Preprocessing

```python
#Pr√©-traitement simple
Lower_Data_train = []
for w in Data_train:
    w = w.lower() #Mise en minuscules
    w = re.sub("<(br|html|body|tr|td|th|table|div|span|a).*>", "", w) #Supprimer les tags html
    Lower_Data_train.append(w)
```
