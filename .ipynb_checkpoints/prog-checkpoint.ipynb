## Choix du dataset d'entrainement

```python
Data_classification = Data_train_clean[:]
Labels_classification = Labels_train[:]
```

## Classifieur TF/IDF + Multinomial Naive Bayes

```python
print("Training Multinomial NB")
pipe_tfidf_mnb = Pipeline([
    ('tfvect', TfidfVectorizer(ngram_range=(1,2), use_idf=True, smooth_idf=True)),
    ('mnb', MultinomialNB(alpha=0.85, fit_prior=true))
    #('bagg', BaggingClassifier(KNeighborsClassifier(), n_estimators=10))
])
clf_tfidf_mnb = pipe_tfidf_mnb.fit(Data_classification, Labels_classification)
print("Done.")
```

```python
joblib.dump(clf_tfidf_mnb, retrieve_name(clf_tfidf_mnb) + '.pkl') # dump le classifieur sur le disque
```

## Classifieur  TF/IDF + Logistic Regression
Individuellement, c'est le meilleur testé

```python
print("Training Logistic Regression")
pipe_tfidf_lr = Pipeline([
    ('tfvect', TfidfVectorizer(max_features = 64000, 
                             ngram_range = (1, 3), 
                             sublinear_tf = True)),
    ('lr', LogisticRegression(C = 7,
                              max_iter = 100,
                              fit_intercept = true,
                              solver = "liblinear",
                             ))
])
clf_tfidf_lr = pipe_tfidf_lr.fit(Data_classification, Labels_classification)
print("Done.")
```

```python
joblib.dump(clf_tfidf_lr, retrieve_name(clf_tfidf_lr) + '.pkl') # dump le classifieur sur le disque
```

## Classifieur TF/IDF + Stochastic Gradient Descent

```python
print("Training Stochastic Gradient Descent")
pipe_tfidf_sgd = Pipeline([
    ('tfvect', TfidfVectorizer(max_features = 150000, 
                             ngram_range = (1, 3), 
                             sublinear_tf = True)),
    ('sgd', SGDClassifier(random_state = 0, 
                shuffle = True, 
                loss = 'modified_huber', 
                alpha = 8e-05,
                tol = None,
                max_iter = 5,
                learning_rate = "optimal"#,
                #eta0 = 1.0
                )),
])
clf_tfidf_sgd = pipe_tfidf_sgd.fit(Data_classification, Labels_classification)
print("Done.")
```

```python
joblib.dump(clf_tfidf_sgd, retrieve_name(clf_tfidf_sgd) + '.pkl') # dump le classifieur sur le disque
```

## Classifieur TF/IDF + KNN

```python
print("Training KNN")
pipe_tfidf_knn = Pipeline([
    ('tfvect', TfidfVectorizer(#max_features = 64000, 
                             ngram_range = (1, 3), 
                             sublinear_tf = True)),
    ('knn', KNeighborsClassifier(n_neighbors=10))
])
clf_tfidf_knn = pipe_tfidf_knn.fit(Data_classification, Labels_classification)
print("Done.")
```

```python
joblib.dump(clf_tfidf_knn, retrieve_name(clf_tfidf_knn) + '.pkl') # dump le classifieur sur le disque
```

## Recherche d'optimisation de modèle

### Principe de vote à la majorité pour faire des prédictions (VotingClassifier)
-----

Cet algorithme nous permet de combiner plusieurs algorithmes de classification associés à des valeurs de confiance (poids) -> Meta-classifieur qui équilibre les faiblesses des classifieurs individuels

```python
print("Training Voting Classifier")
clf1 = pipe_tfidf_mnb
clf2 = pipe_tfidf_lr
clf3 = pipe_tfidf_sgd
clf4 = pipe_tfidf_knn
pipe_voting = Pipeline([
    ('voting', VotingClassifier(estimators=[
        ('mnb', clf1), 
        ('lr', clf2),
        ('sgd', clf3),
        ('knn', clf4)
    ], 
    voting = 'soft',
    weights=[4,7,1,2]))
])
clf_voting = pipe_voting.fit(X_classification, y_classification)
print("Done.")
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=true,
                          title='Confusion matrix',
                          cmap=plt.cm.winter):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)
    fig = plt.figure(figsize=(12,7))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center", fontsize=25,
                 color="white" if cm[i, j] < thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    filename = 'ConfusionMatrix.png'

    #plt.show()
    fig.tight_layout()
    fig.savefig("./visualisation/" + filename)

plot_confusion_matrix(results["confusion_matrix"], [-1, 1], normalize=true)
```
