```python
Data_classification = Data_train_clean[:]
Labels_classification = Labels_train[:]
```

## Classifieur Multinomial Naive Bayes (NB)

```python
print("Training Multinomial NB")
pipe_tfidf_mnb = Pipeline([
    ('tfvect', TfidfVectorizer(ngram_range=(1,2), use_idf=True, smooth_idf=True)),
    ('mnb', MultinomialNB(alpha=0.85, fit_prior=True))
])
clf_tfidf_mnb = pipe_tfidf_mnb.fit(Data_classification, Labels_classification)
print("Terminé.")
```

```python
joblib.dump(clf_tfidf_mnb, retrieve_name(clf_tfidf_mnb) + '.pkl') # sauvegarde le classifieur
```

## Classifieur Stochastic Gradient Descent

```python
print("Training Stochastic Gradient Descent")
pipe_tfidf_sgd = Pipeline([
    ('tfvect', TfidfVectorizer(max_features = 150000, 
                             ngram_range = (1, 3), 
                             sublinear_tf = True)),
    ('sgd', SGDClassifier(random_state = 0, 
                shuffle = True, 
                loss = 'modified_huber', 
                alpha = 8e-05,
                tol = None,
                max_iter = 5,
                learning_rate = "optimal"#,
                #eta0 = 1.0
                )),
])
clf_tfidf_sgd = pipe_tfidf_sgd.fit(Data_classification, Labels_classification)
print("Terminé.")
```

```python
joblib.dump(clf_tfidf_sgd, retrieve_name(clf_tfidf_sgd) + '.pkl') # sauvegarde le classifieur
```

## Classifieur KNN

```python
print("Training KNN")
pipe_tfidf_knn = Pipeline([
    ('tfvect', TfidfVectorizer(#max_features = 64000, 
                             ngram_range = (1, 3), 
                             sublinear_tf = True)),
    ('knn', KNeighborsClassifier(n_neighbors=10))
])
clf_tfidf_knn = pipe_tfidf_knn.fit(Data_classification, Labels_classification)
print("Terminé.")
```

```python
joblib.dump(clf_tfidf_knn, retrieve_name(clf_tfidf_knn) + '.pkl') # sauvegarde le classifieur
```

### Roc curve

```python
def get_roc_curve(pred_proba, X, y):
    #pred_proba = model.predict_proba(X)[:, 1]
    fpr, tpr, _ = roc_curve(y, pred_proba)
    return fpr, tpr

roc_svm = get_roc_curve(pred_proba, Data_analysis, Labels_analysis)

fpr, tpr = roc_svm
fig = plt.figure(figsize=(12,7))
plt.plot(fpr, tpr, color="red")
plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')
plt.xlim([-0.004, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('true Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Roc curve')
filename = 'ROC_AUC.png'

plt.show()
fig.tight_layout()
fig.savefig("./visualisation/" + filename)
```

### Confusion Matrix

```python
def plot_confusion_matrix(cm, classes,
                          normalize=True,
                          title='Confusion matrix',
                          cmap=plt.cm.winter):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)
    fig = plt.figure(figsize=(12,7))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center", fontsize=25,
                 color="white" if cm[i, j] < thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    filename = 'ConfusionMatrix.png'

    #plt.show()
    fig.tight_layout()
    fig.savefig("./visualisation/" + filename)

plot_confusion_matrix(results["confusion_matrix"], [-1, 1], normalize=False)
```
